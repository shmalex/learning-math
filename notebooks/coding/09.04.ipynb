Say you have an n-by-n matrix of elements that are sorted in ascending order both in the columns and rows of the matrix.

Return the k-th smallest element of the matrix.

For example, consider the matrix below:
```
1 4 7
3 5 9 
6 8 11
```
if k=4, then return 5.

````python
a= [[1,4,7],[3,5,9],[6,8,11]]
````

````python
def get_matrix(n):
    a = np.random.randint(0, 100, size=(n,n))
    a = np.sort(a, axis=1)
    a = a.transpose()
    a.sort(axis=1, )
    return a
````

````python
from heapq import heappush, heappop
````

````python
def find_kthsmallest_v1(a, k):
    _i = 0
    n = len(a)
    minarr = []
    for i in range(n):
        for j in range(n):
            _i +=1
            heappush(minarr, (a[i][j], a[i][j]))
    res = a[0][0]
    for _ in range(k):
        _i+=1
        res, _ = heappop(minarr)
    return res, _i
````

````python
def find_kthsmallest_v2(a, k):
    _i = 0 # operations counter
    n = len(a)
    minarr = []
    maxarr = a[0][0]
    res = a[0][0]
    for i in range(n):
        for j in range(n):
            _i +=1
            if len(minarr)>=k:
                if maxarr < a[i][j]: break
                else:
                    heappush(minarr, (a[i][j], a[i][j]))
            else:
                maxarr = max(maxarr, a[i][j])
                heappush(minarr, (a[i][j], a[i][j]))
    
    for _ in range(k):
        _i+=1
        res, _ = heappop(minarr)

#     print(res, _i)
    return res, _i
````

````python
results = []
# test
for _ in range(100000):
    k = np.random.randint(0, 100)
    a = get_matrix(10)
    res1, i1 = find_kthsmallest_v1(a, k)
    res2, i2 = find_kthsmallest_v2(a, k)
    results.append([k, res1, i1, res2, i2])
results = np.array(results)
#     print(k, res1, res2, res1==res2, i1, i2)
````
