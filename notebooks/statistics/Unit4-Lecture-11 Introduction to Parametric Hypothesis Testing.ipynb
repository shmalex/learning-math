# 3. Null and alternative hypotheses


Какое ожидание в скорой ER - Emergency Room

что в среднем в скорой ожидают помощи 30 минут.

А кто-то из пациентов говорит - там на много дольше нужно ждать

Мы как статистика пошли проверять это жалобы.

мы пошли собрали выборку $X_1,\dots,X_n$ какого-то значительного размера и все наблюдения независимы.

Из эти данных мы можем вычислисть $\mathbb{E}[X_1]=\mu$

Мы хотим знать если на самом деле $\mu > 30$


В терминах статистической гипотезы это перводится в термины 
$$H_0: \mu \le 30\quad \text{null hypothesis}$$
$$H_1: \mu > 30\quad \text{alternative hypothesis}$$

### статистическая формализация


Предположим выборку $X_1,\dots,X_n$ iid случайная переменная и в нашем примере это время ожидания приема.

Статистическая модель $(E,(\mathbb{P}_\theta)_{\theta\in\Theta})$ - где $(\mathbb{P}_\theta)$ это семейство распределений индексированная абстрактным пространством $\Theta$. В этом пространстве находится одна единственная истинная $\theta^*$. Мы разделим  это пространство на две **партиции/части**

**Партиция** - это два непересекающихся набора, объединение которых составлять весь $\Theta$:

Пуст $\Theta_0$ и $\Theta_1$ являются партициями/частями $\Theta$

И вопрос который мы ставим - где находится $\theta^*$ в первой или во второй партиции? Этот вопрос выливается в нулевую и альтернативную гипотизы, каждая из которых заявляет что иcтинный параметр находится в их партиции.

Конечно на нашем рисунке нам птичка на хвосте принесла информацию где находится параметр, но на практике такого не бывает.

Учтем две гипотезы 
${{\begin{cases}
H_0:\theta\in\Theta_o\\
H_1:\theta\in\Theta_1\\
\end{cases}}}$

Где $H_0$ это Нулевая гипотеза и $H_1$ это альтернативная гипотеза. У них разные названия, потому что они играют разные роли, ассиметричные роли, в статистических тестах гипотез

Мы говорим что тестируем $H_0$ против $H_1$

**ВАЖНО - чтобы наши гипотезы создавали две партиции**

# 4. Different types of hypotheses

Будем обсуждать типы гипотез. в целом все вроде бы похоже, но природа у них разная.

введем математическую нотацию для гипотиз

Для $k=0\,(H_0)$ или $k=1\,(H_1)$ получается что это $k$ это как ссылна на гипотезу $H_k$ таким образом мы просто обозначаем через $k$ сразу и $H_0$ и $H_1$.

- $\Theta_k$ это простая гипотиза если $\Theta_k=\{\theta_k\}$- тоесть это одно число, или набор из одного элемента.
- $\Theta_k$ - копмозитная гипотеза если $\Theta_k$ имеел след форму

$$\Theta_k=\{\theta:\theta>\theta_k\},\quad \Theta_k=\{\theta:\theta<\theta_k\}, \quad\Theta_k=\{\theta:\theta\neq\theta_k\}, $$

Первый варианта это значит $\Theta_k$ состоит из набора в котором $\theta$ больше чем $\theta_k$ из простой гипотезы. тоесть это набор множестве чисел которые больше, меньше или не равно $\theta_k$.

Это пример трех композитных гепотих которые у нас могут быть.


В заквисимости какая у нас композиция - от этого зависит какой у нас тест **"One sided" "Two Sided"**


${{
\begin{align}
\quad\text{Two-sided}\quad\quad\quad&\quad\quad\quad\quad\text{One-sided}\\
\begin{cases}
H_0:\theta=\Theta_0\\
H_1:\theta\neq\Theta_0\\
\end{cases}\quad\quad&
\begin{cases}
H_0:\theta\leq\Theta_0\\
H_1:\theta>\Theta_0\\
\end{cases}
\begin{cases}
H_0:\theta\geq\Theta_o\\
H_1:\theta<\Theta_0\\
\end{cases}
\end{align}}}$



Разберем пример.

1) Время ожидания ER
$$\begin{cases}
H_0:\mu\le30\\
H_1:\mu>30\\
\end{cases}$$

- Pассмотрим к какому типу относятся гипотезы\
В это случае $H_0$ - так как $\mu$ равно всему что меньше 30. Значит она принадлежит набору из множества чисел - **композитная**\
В это случае $H_1$ - так как $\mu$ тоже  **композитная** так как ей соотв. набор из множества значений

- "One sided" или "Two Sided" ?\
Так как $H_1$ говорит (хотя достаточно посмотреть на любоую из них) нам что $\mu$ находится ТОЛЬКО с одной стороны то это **"One sided"** тест

2) Kiss example
$$\begin{cases}
H_0:p=.5\\
H_1:p\neq.5\\
\end{cases}$$

- Рассмотрим к какому типу относятся гипотезы\
В это случае $H_0$ - так как $p$ равно конкретному числу, тогда это - **простая** гипотеза
В это случае $H_1$ - так как $p$ равна всему чему угодно кроме конкретного одного числа - **композитная** гипотеза.

- "One sided" или "Two Sided" ?\
так как $H_1$ равна чемуто по обе стороны от 0.5 - то это считается **Tow Sided**


# 5. Statistical modelling

Это реальная тема из статистики. Мы должны научиться конвертить проблемы в статистические формулировки.

Как работают клинические испытания -  есть **котрольные группы**, которым получают плацебо.
Есть **тестовые группы** которые получают лекарство. 

К примеру мы дали группе людей лекарство которое должно снизить их LDL среди пациентов у которых LDL выше 200 mg/dL.

### Разберемся с нотацией.

Пусть $\mu_d>0$ (d - drug) обозначает ожидаемое уменьшение LDL для пациентов которые принимали лекарство.

Пусть $\mu_c>0$ (c - control) обозначает ожидаемое уменьшение LDL для пациентов которые принимали плацебо.

Hypothesis testing problem:
$$
\begin{cases}
H_0: \mu_d\le\mu_c\,\text{читается как "не лучше"}\\
H_1: \mu_d>\mu_c
\end{cases}
$$


Мы наблюдаем две независымые выборки

- $X_1,\dots,X_n\stackrel{iid}{\sim }\mathcal{N}(\mu_d,\sigma^2_d) $ для тестовой группы
- $X_1,\dots,X_n\stackrel{iid}{\sim }\mathcal{N}(\mu_c,\sigma^2_c) $ для контрольной группы

такой тест называется "двух выборочный тест" (a/b testing)

Это двух-сторонний текст. 

План\
1 - рассмотрим ситуаци (группы)\
2 - определим гипотезы\
3 - определим выборки - как собираем данные\
4 - определим скольки сторонние тест

### Two Sample vs. One Sample Tests
### Тесты с двумя выборками против тестов с одной выборкой

В **Одна выборка** - это тест гипотетезы того что один **не известный** параметр $\mu$ сравнивается с **известным** значением эталонного параметра.

3) $\Theta_0$ и $\Theta_1$ не составляют все прострство выборки  $\Theta$. потому что могуть быть тест в которых пространство выборки может быть задано $\Theta =\{ (\mu _1,\mu _2):\mu _1\in \mathbb {R},\, \mu _2\in \mathbb {R}\} =\mathbb {R}^2$. Вне зависимости от данных, мы никогда не **"принимаем нулевую гипотезу"**, мы либо ее **Отвергаем** либо не **не в состоянии отклонить** ее. 

# 7. Test

Что же такое Test?

Тест это статистика $\psi\in\{0,1\}$ которая явно не зависит от неизвестной величины. (но профессор говорит что иногда бывает что зависит). Тест это еще Оценщик, функция от данных, который может быть равен только 0 или 1, бинарные значение.

$$\begin{cases}
\text{if} \, \psi=0, H_0 \text{is not rejected}\\
\text{if} \, \psi=1, H_1 \text{is rejected}
\end{cases}$$

Если у нас есть выборка $X_1, \ldots , X_ n \stackrel{iid}{\sim } \text {Ber}(p^*)$ для какого-то параметра $p^* \in (0,1)$. Пусть $(\{ 0,1\} , \{ \mathbf{P}_ p\} _{p \in (0,1)})$ это стат модель, где $\mathbf{P}_ p = \text {Ber}(p)$\

Где $\Theta _0, \Theta _1 \subset (0,1)$ и $\Theta _0$ не пересекается с $\Theta _1$

Мы проводим статистический тест
$\quad\psi :\{ 0,1\} ^ n \to \{ 0, 1\}$\
это значит что у нас есть $n$ кортежей каждый из которых может принимать значение из набора $\{0, 1\}$, и посредством теста этих картежей мы "мапим" результат на каржет $\{0, 1\}$. Эта функция теста бере в качестве входного параметра выборку $(X_1, \ldots , X_ n)$. И если $\psi (X_1, \ldots , X_ n) = 1$ значит мы **Reject the null $H_0$** или **fail to reject null**



По факту Тест это индикаторная переменная $\psi=\mathbb{1}\{R\}$ которая обозначает событие называемое **Rejection Region**.
$\psi=\mathbb{1}(\psi=1)$ это тавтология. То пространство где наступает событие $\psi=1$ это регион **Rejection Region**

# 9. Level and asymptotic level

У нас есть две ошибки, которые соревнуются друг с другом, которые описываются $\beta(\theta)$ функцией. Их соревнование формирует компромис между двумя ошибками, который называется парадигма Neyman-Pearson.

На примере ER мы пытаемся определить как выбрать C\
$\quad\quad\quad$
$\begin{cases}
H_0:\mu \le 30\\
H_1:\mu > 30 \quad 
\end{cases}$$\psi=\mathbb{1}\{\bar{X}_n>C\}$

Суть в том что мы не хотим чтобы обе ошибки были наименьшими.
- Чтобы вероятность ошибки **первого** типа $\to 0$, взять $C\to+\infty$
- Чтобы вероятность ошибки **второго** типа $\to 0$, взять $C\to-\infty$

**$С$  это про X - в одиница данных**

и не может чтобы $C$ - была одновременно двух значение.
Это называется **multy objective optimization**

The Neyman-pearson paradigm - давайте решим что один тип ошибки важнее чем другой типа, и они решили что более важная - это Type 1.
- Убедитесь что вероятность ошибки первого типа **ограничена** $\mathbb{P}[\text{Type 1 error}]\le\alpha$ (Пр: $\alpha=5\%,1\%$)
- **Минимизировать** вероятност ошибки второго тпаи

Мы будем проводить такие тесты при которых вероятность ошибки первого рода не более $\alpha$. Исходя из этого мы хотим минимизировать ошибку второго рода как предмет этого ограничения.

Тогда эта $\alpha\in[0,1)$ и называется **уровнем теста, level of a test**.

**$\alpha$ - это вероятности**

#### Как же мы определяем уровень?

Для какиех $\theta\in\Theta_0$ мы должны считать  $\mathbb{P}[\psi=1]$ - Мы хотим иммунитета от "Type 1 error" не важно для какого либо $\theta$

- Тест $\psi$ уровня $\alpha$ если\
$$\mathbb{P}[\psi=1]\le\alpha, \forall \theta \in \Theta_0$$
Это тоже само что сказать что для максимальная вероятность первой ошибки для любых $\theta$ не больее $\alpha$
$$\max_{\theta\in\Theta_0}\mathbb{P}_\theta[\psi=1]\le\alpha$$

$$\alpha=\sup_{\theta\in\Theta_0}\beta(\theta)$$
**$\alpha$ - Это верхняя граница при которой мы отклоним нулевую гипотезу** означает, что мы ищем самую большую (верхнюю границу) вероятность правильно отвергнуть нулевую гипотезу среди всех возможных значений $\theta$, при которых нулевая гипотеза верна. При этом важно не забывать что $\alpha$ это площадь под кривой, а не значение $X$



Мы точно не знаем распределения средней, для которых мы обычно будет считать вероятность теста. Однако если использовать CLT мы сможем подсчитать асимптотичское распределение правильно центрированной и мастабированной версии $\bar{X}_n$ тогда мы сможет говорить об **Asymptotic level $\alpha$**

- Тест $\psi=\psi_n$ имеет асимптотический уровень $\alpha$ есть\
$$\lim_{n\to\infty}\max_{\theta\in\Theta_0}\mathbb{P}_\theta[\psi_n=1]\le\alpha, \forall \theta \in \Theta_0$$

То есть у нас есть интервал в который входит $\theta$ и у нас нет причин верить в то что $\theta$ не принадлежит этому отрезку

У нас возникает 2 сценария, 



- первый $\theta_0$ принадлежит $[A,B]$ если это так тогда у нас нет причин не верить что $\theta_0$ не является истенным значением для $\theta$ и у нас нет явных доказтельств против этой конкретной $\theta_0$, у нас также нет доказательств против любого значение $\theta_0$ внутри этого интервала. И если случиться так что $\theta_0$ будет в нутри этого интервала то мы **не смогли отклонить нулевую гипотезу**



- $\theta_0$ за пределами $[A,B]$ тогда у это значение не може быть правдоподобным значение для $\theta$, оно даже не принадлежит $1-\alpha$ CI для $\theta$

**как записать это в математических терминах?**

$\psi=\mathbb{1}\{\theta_0 \notin[A,B]\}$ - **Rejection region**

**как определить уровень**

$\mathbb{P}_{\theta_0}[\psi=1]=\mathbb{P}_{\theta_0}[\theta_0\notin I]$
По определению мы не знаем как контролировать вероятность что $\theta$ не принадлежит к CI, но по определению (1) мы знаем как контролировать вероятность что $\theta$ принадлежит CI. И знаем что это равно как минимум $1-\alpha$. И так применяя правино комплементарности мы можем переписать как 

$=1-\mathbb{P}_\theta(\theta\in[A,B])\ge1-(1-\alpha)=\alpha$

**Получается что Ошибка Первого рода контролируется посредством $\alpha$**

### Пример Kiss

мы тестируем
$$\begin{cases}
H_0:p=.5\\
H_1:p\neq.5
\end{cases}$$

Мы собрали выборку $R_1,\dots,R_n\stackrel{iid}{\sim }\text{Ber}(p)$
- вспомним что $I_\text{conserv}=[\bar{R}_n-\frac{1.96}{2\sqrt{n}},\bar{R}_n+\frac{1.96}{2\sqrt{n}}]$, где $2=\sqrt{\frac{1}{2}(1-\frac{1}{2})}$это была $\sigma$. это CI asymptotic level $1-\alpha$ для $p$ где

- Тест
$$\psi=\mathbb{1}\{0.5\notin I_\text{conserv}\}$$
Соотвественно мы отклоним нулевую гипотезу если $0.5\notin I_\text{conserv}$

$\text{if}\,\alpha=5\%$ вспомним что мы построили $I_\text{conserv}=[0.56,0.73]$ в прошлых лекциях. и так как 0.5 не входит в этот интервал, значит мы отклоним этот тест.

- У нас есть (это асимптотический уровень)\
$\lim_{n\to\infty}\mathbb{P}_{0.5}[\psi=1]=1-lim_{n\to\infty}\mathbb{P}_{0.5}[0.5\in I_{conserv}]\le1-(1-\alpha)=\alpha$
- По этому $\psi$ это тест с асимптотическим уровнем $\alpha$

Тут профессор говорит о том что мы знаем асимптотический CI

**(i) Мы не относимся к неизвестному параметру как к случайной переменной. Не имеет смысла обсуждать его распределение.**

# 12. P-values

Рассмотрим наш пример с Kiss

$I_\text{conserv} = [\bar{R}_n \pm \frac{q_{\alpha/2}}{2\sqrt{n}}]$

С нашими данными $I_\text{conserv}=[0.56, 0.73]$ мы отвергаем Нулевую гипотезу на уровни **5%**

| $\alpha$ | $q_{\alpha/2}$| $I_{conserv}$ |      Decision     |
|:--------:|:-------------:|:-------------:|:-----------------:|
|    10%   |      1.64     |  [0.57, 0.72] |      Reject       |
|     5%   |      1.96     |  [0.56, 0.73] |      Reject       |
|     1%   |      2.576    |  [0.52, 0.77] |      Reject       |
|     .1%  |      3.29     | [0.497, 0.79] | Failed to Reject  |
|     .01% |      3.89     |  [0.47, 0.82] | Failed to Reject  |

мы видем что с уменьшением $\alpha$ - тоесть мы делаем его более уверенным, у нас расширяется CI и CI **начинает включать** в себя значение $H_0=0.5$ и мы уже **не может отвергнуть нулевую гипотезу**. Если мы дойдем до нуля, то мы никогда не будем отвергать нуль гипотезу.

Что важно это значение $\alpha$ при котором мы "переключаемся" между "Reject" к "Failed to reject". Мы видим что это значение между $[1\%, .1\%]$. нам интересно знать что это за значение? Оно называется **$p-\text{value}$**

## $p-value$

#### Определение

(Асимптотическое) $p-value$ теста $\psi$ это наименьший (асимптотический) уровень $\alpha$ при котором тест $\psi$ отвергает гипотезу $H_0$

Как только мы знаем чему оно равно.

#### Golden Rule

$p-\text{value} \le \alpha \Leftrightarrow H_0$ **отвергнута тестом $\psi$** , на (асимптотическом) уровне $\alpha$

$p-\text{value} > \alpha \Leftrightarrow H_0$ **не отвернута тестом $\psi$** , на (асимптотическом) уровне $\alpha$

#### Kiss

нам нужно найти $\alpha_0$ такой что $\bar{R}_n-\dfrac{q_{\frac{\alpha_0}{2}}}{2\sqrt{n}}=0.5$ (Lower bound)

if $\bar{R}_n=.645, n=124$ мы получим $q_{\frac{\alpha_0}{2}}=3.23$

Чтобы найти $\alpha_0$

$\dfrac{\alpha}{2}=\mathbb{P}[Z>q_{\dfrac{\alpha_0}{2}}]=\mathbb{P}[Z>3.23]=1-0.9994=0.06\%\rightarrow\,\alpha_0=0.12\%$


Где $Z\sim\mathcal{N}(0,1)$ и $\mathbb{P}(Z\le 3.24)=0.9994 (взято из таблици)
