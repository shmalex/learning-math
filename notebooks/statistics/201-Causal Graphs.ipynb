 # Коллайдерная переменная

Коллайдерная переменная (collider) в статистике возникает в случае, когда две независимые переменные (A и B) влияют на третью переменную (C). Пример такой модели можно найти в эпидемиологии, социологии или экономике. 



Причинные переменные $A$ и $B$, которые влияют на коллайдер $C$, не обязательно связаны между собой.
Примеры таких перменных
- A: Образование.
- B: Доход.
- C: Стресс.

Или 

- A: Уровень физической активности
- B: Курение
- C: Заболевание сердца

**Модель:**
1. Физическая активность и курение могут оба влиять на вероятность развития заболевания сердца.
1. Физическая активность и курение считаются независимыми факторами, не влияющими друг на друга напрямую.
1. Заболевание сердца может быть обусловлено либо низким уровнем физической активности, либо курением, либо сочетанием этих факторов.

**Интерпретация:**
- Физическая активность (A) влияет на риск заболевания сердца (C).
- Курение (B) также влияет на риск заболевания сердца (C).
- Физическая активность и курение не зависят друг от друга (нет прямой зависимости между A и B).

В данном случае, физическая активность (A) и курение (B) влияют на вероятность заболевания сердца (C). Или - Уровень стресса (C) определяется как результат комбинации уровня образования (A) и уровня дохода (B),  

Схематично это можно изобразить так:

**$$A\,\,\to\,\,C\,\,\leftarrow\,\,B$$**


Если они не связаны между собой, **коллайдер не защищён**. 

 ### Фиксация коллайдера

Коллайдер — это переменная, на которую влияют две другие независимые переменные. Эти две переменные остаются независимыми, если не учитывать значение коллайдера.

**Фиксация коллайдера** означает, что мы выбираем подмножество данных, где коллайдер имеет **определенное значение или диапазон значений**. Например, выбираем только людей с **заболеванием сердца ($C$ = 1)**.
#### Появление ложной зависимости:

Если мы фиксируем значение коллайдера, между двумя независимыми переменными, влияющими на него, **может возникнуть ложная зависимость**.
Например, среди людей с заболеванием сердца может казаться, что физическая активность и курение связаны, хотя в общей популяции такой зависимости нет.

#### Отсутствие зависимости без фиксации:

Если коллайдер не фиксирован, то есть мы рассматриваем всю популяцию без выборочного отбора по значению коллайдера, две независимые переменные останутся независимыми.
В этом случае физическая активность и курение остаются несвязанными.

### Пример:

**Без фиксации коллайдера:**
- Рассматриваем всю популяцию.
- A и B независимы.
- Мы не видим связи между физической активностью и курением.

**С фиксацией коллайдера:**
- Рассматриваем только людей с заболеванием сердца (C = 1).
- В этом подмножестве может появиться видимая зависимость между A и B.
- Внешне будет казаться, что физическая активность и курение связаны, чтобы объяснить наличие заболевания сердца.

 ### Не защищенность коллайдера и причинно-следственные связи

**Почему термин "не защищён":**
Термин "не защищён" акцентирует внимание на том, что мы не защищаем независимость переменных A и B. Когда коллайдер **"не защищён" (то есть его значение фиксировано (С=1) )**, между A и B может возникнуть ложная зависимость.

Можно сказать, что коллайдер влияет на **видимость зависимости между двумя переменными**, но не объясняет причинно-следственные связи. Это важно учитывать, так как:

**Коллайдер влияет на видимость зависимости:**

- Коллайдер может создавать ложную зависимость между переменными при условной фиксации его значения.
- Без фиксации коллайдера этой зависимости не будет видно, даже если существует косвенная связь через коллайдер.

**Причинно-следственные связи:**
- Причинно-следственные связи описывают, как изменение одной переменной (причины) приводит к изменению другой переменной (следствия).
- **Коллайдер сам по себе не является объяснением причинно-следственной связи, а скорее фактором**, который влияет на то, видим ли мы ложную зависимость между независимыми переменными.

#### Пример 
**Допустим, у нас есть три переменные:**

- A: Уровень физической активности.
- B: Курение.
- C: Заболевание сердца.

Причинно-следственные связи здесь таковы:

- Физическая активность (A) влияет на вероятность заболевания сердца (C).
- Курение (B) также влияет на вероятность заболевания сердца (C).

**Без фиксации коллайдера (C) (коллайдер защищён):**
- Если мы рассматриваем общую популяцию, мы видим только прямые причинно-следственные связи между A и C, а также между B и C.
- Между A и B прямой зависимости нет, так как коллайдер C "блокирует" эту связь.

**С фиксацией коллайдера $C$ (коллайдер не защищён):**
- Если мы фиксируем C (например, выбираем только людей с заболеванием сердца), между A и B может появиться ложная зависимость.
- Эта зависимость не является истинной причинно-следственной связью, а лишь результатом условной фиксации коллайдера.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Генерация данных
np.random.seed(42)
n_samples = 10000

# Независимые переменные A и B
A = np.random.normal(loc=0, scale=1, size=n_samples)
B = np.random.normal(loc=0, scale=1, size=n_samples)

# Коллайдер C зависит от A и B
C = 0.5 * A + 0.5 * B + np.random.normal(loc=0, scale=1, size=n_samples)

# Создание DataFrame
df = pd.DataFrame({'A': A, 'B': B, 'C': C})

# Расчет корреляции в общей популяции
corr_all, _ = pearsonr(df['A'], df['B'])

# Фиксация значения коллайдера C
fixed_C_df = df[np.abs(df['C']) < 0.1]

# Расчет корреляции при фиксированном значении C
corr_fixed_C, _ = pearsonr(fixed_C_df['A'], fixed_C_df['B'])

# График зависимости между A и B в общей популяции
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(df['A'], df['B'], alpha=0.1)
plt.title(f'Общая популяция: Зависимость между A и B\nКорреляция: {corr_all:.2f}')
plt.xlabel('A (Физическая активность)')
plt.ylabel('B (Курение)')

# График зависимости между A и B при фиксированном значении C
plt.subplot(1, 2, 2)
plt.scatter(fixed_C_df['A'], fixed_C_df['B'], alpha=0.1, color='orange')
plt.title(f'Фиксация коллайдера: Зависимость между A и B при фиксированном C\nКорреляция: {corr_fixed_C:.2f}')
plt.xlabel('A (Физическая активность)')
plt.ylabel('B (Курение)')

plt.tight_layout()
plt.show()
```


Причинные переменные, которые влияют на коллайдер, не обязательно связаны между собой. Если они не связаны между собой, **коллайдер не защищён**.
