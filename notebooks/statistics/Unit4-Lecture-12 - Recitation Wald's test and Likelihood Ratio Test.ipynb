# Recitation problem statement

Suppose we have a dataset $X_1,\dots,X_n$ that is i.i.d. $f_\theta (x) = \theta x^{-\theta - 1} \mathbf{1}_{x > 1}$, $\theta>0$ .

1. Calculate the MLE and Fisher Information for $\theta$.
2. Give the likelihood ratio test with level $\alpha=0.05$ to test if $\theta=2$.
3. Give Wald's test with level $\alpha=0.05$ to test if $\theta=2$.
4. Assume $n=100$ and $\hat\theta=2.45$. Compute the asymptotic -values of both tests.

#  1. Review: Wald's Test and Likelihood Ratio Test

тут мы будет рассматривать тестирование гипотез используя два MLE теста
1) Wald's test - основанный на асимптотически нормальной аппроксимации к MLE оценке.
2) Likelihood ration test - смотрит на сколько правдаподобно что $\theta\in\Theta_0$ чем $\theta\notin\Theta_0$

$H_0: \theta=\theta_0$\
$H_1: \theta\neq\theta_0$

У нас есть данные $X_1, \dots,X_2\sim f(x,\theta)=\theta x ^{-\theta-1}, x\geq1, \theta>0$ в этой модели мы хотим оценить $\theta$ и проверить гипотезу $H_0$. Pareto distribution

После этого мы проверим другую гипотезу\
$H_0:\theta=2$\
$H_1:\theta\neq2$

после того как мы выведем тесты, мы проведем их:

$\hat\theta_{text{MLE}}=2.45, n=100$



## Вспомним что такое Wald's Test.

Walds test основывается на асимптотически нормальном приближение к MLE. А асимптотически нормальное приближение к MLE говорит что 
$\sqrt{n}(\hat\theta_\text{MLE}-\theta)\xrightarrow{(d)}\mathcal{N}(0, I^{-1}(\theta))$

Мы можем умножить обе стороны на $\sqrt{I(\theta)}$ и мы получим что мы "взвесили" обе части,и тогда wald's test сойдется к нормальному распределенеию.

$\sqrt{n}\sqrt{I(\theta)}(\hat\theta_\text{MLE}-\theta)\xrightarrow{(d)}\mathcal{N}(0,1)$

А теперь мы возведем во 2 степень и наше выражение с лево сойдется к\
$n I(\theta)(\hat\theta_\text{MLE}-\theta)^2\xrightarrow{(d)}\chi_1^2$

и еще что обычно делают, мы не не хотим много параметров мы используем метод "plug-in" и получаем\
$$n I(\hat\theta)(\hat\theta_\text{MLE}-\theta)^2\xrightarrow{(d)}\chi_1^2$$
Это и есть Wald's test, мы будем использовать эту статистику и ее асимптотическое распределение для тестирования гипотез.

## Likelihoo ratio test

$\dfrac{\text{правдоподобие данных при альтернативной гипотезе}}{\text{правдоподобие данных при нулевой гипотезе}}$

$\dfrac{L(X_1, \dots,X_n;\Theta)}{L(X_1,\dots,X_n;\Theta_0)}$

Так как при альтернативной гипотезе мы тестируем очень много разных значений $\theta$ тогда нам нужно выбрать те что дадут максимум правдоподобия.

$\dfrac{\max_{\theta\in\Theta_1}\text{правдоподобие данных при альтернативной гипотезе}}{\text{правдоподобие данных при нулевой гипотезе}}$

$\dfrac{\max_{\theta\in\Theta_1}L(X_1, \dots,X_n;\Theta)}{L(X_1,\dots,X_n;\Theta_0)}$


Соотв. мы отвергнем нулевую гипотезу если наша пропрция будет больше чем константа, определенная значемым уровнем, которая больше 1. 
$\dfrac{\max_{\theta\in\Theta_1}L(X_1, \dots,X_n;\Theta)}{L(X_1,\dots,X_n;\Theta_0)}>c, c>1$

Уточнение, так как пропорция от "Открытого множества", мы вместо $\max$ должны указать $\sup$

$\dfrac{\sup_{\theta\in\Theta_1}L(X_1, \dots,X_n;\Theta)}{L(X_1,\dots,X_n;\Theta_0)}>c, c>1$

И если мы посмотрим на картину в целом - $sup$ от все возможных вариантом $\theta$ это и будет MLE

$\dfrac{L(X_1, \dots,X_n;\hat\Theta^{\text{MLE}})}{L(X_1,\dots,X_n;\Theta_0)}>c, c>1$

Возмем логарифм и умножем на 2, (это делается чтобы применить теорию Wilk)

$2(\log L(X_1, \dots,X_n;\hat\Theta^{\text{MLE}}) - \log L(X_1,\dots,X_n;\Theta_0)>\log(c), c>1$

Применяем Wilk's теорию - и это значит то статистика выше асимптотически сойдется в распределении к $\chi_1^2$

$$2(\log L(X_1, \dots,X_n;\hat\Theta^{\text{MLE}}) - \log L(X_1,\dots,X_n;\Theta_0)\xrightarrow{(d)}\chi^2_1$$

Мы получили два теста, которые сходятся к $\chi^2$ распределению,
- один это Wald который основывается на квадрате асимптотической нормальной сходимости. 
- второй это likelihood ratio test который сравнивает правдоподобие альтернативной и нулевой гипотиз

# 2

Мы получили два теста, чтобы их провести нам нужно подсчитать "fisher information" и "MLE"

## MLE

$L(X_1,\dots,X_n;\theta)=\prod_{i=1}^n f(x_i;\theta)\
=\prod_{i=1}^n\theta X_i^{-\theta-1}$\
Вынесем $\theta$ за скобки \
$=\theta^n(\prod_{i=1}^nX_i)^{-\theta-1}$

и так мы хотим максимизировать эту функцию, и обычно чтобы это сделать мы берем производную и приравниваем ее к 0. нам будет сложно взять производную этой функции. Вместо этого возьмем логарифм функции.\
$\ell(X_1,\dots,X_n;\theta)=\log\left(\theta^n(\prod_{i=1}^nX_i)^{-\theta-1}\right)\\
=n\log(\theta) +(-\theta-1)(\sum_{i=1}^n\log X_i))$

Теперь возмем производную и прировняем к нулю

$\dfrac{d\ell}{d\theta}=\frac{n}{\theta}-\sum_{i=1}^n\log X_i=0$\
$\theta=\dfrac{n}{\sum_{i=1}^n\log X_i}$

Для проверки того что эта функция является максимизатором, мы возьмем вторую производную и проверим если она вогнутая.

$\dfrac{d^2\ell}{d\theta^2}=-\dfrac{n}{\theta^2}<0$ - да это вогнутая функция. значит функция выше максимайзер, тогда мы можем написать:

$\hat\theta_{\text{MLE}}=\dfrac{n}{\sum_{i=1}^n\log X_i}$



## Fisher information

FI это ожидание минус вторая производная Log Likelihood. Когда мы записываем FI нам нужно взять Log Likelihood от одной копии $X_1$ а не $n$ копий $X$\
$I(\theta)=\mathbb{E}\left(-\dfrac{d^2\ell(X;\theta)}{d\theta^2}\right)=\mathbb{E}\left[-\left(-\dfrac{1}{\theta^2}\right)\right]=\dfrac{1}{\theta^2}$

## Проведем тесты

Итого у нас сейчас есть след:\
$\ell(X_1,\dots,X_n;\theta)=n\log(\theta) +(-\theta-1)(\sum_{i=1}^n\log X_i))$\
$\hat\theta_{\text{MLE}}=\dfrac{n}{\sum_{i=1}^n\log X_i}$\
$I(\theta)=\dfrac{1}{\theta^2}$

### Wald's test

$W=n\dfrac{1}{\hat\theta_{\text{MLE}}^2}\left(\hat\theta_{\text{MLE}}^2-\theta_0\right)^2\xrightarrow{(d)}\chi^2_1$ (under $H_0$)

Что нам говорит это выражение - если значение $W$ очень большое то оно имеет низкую вероятность того что оно было призведено $\chi^2$ распределением. $\chi^2$ всегда положительное и выглядит так.



Суть нашего теста в том что мы отклоним нуль гепотезу, если наша статистика большая. Назвоем левую часть $T_n$

Тогда наш тест будет выглядить как 
$\psi=\mathbb{1}(T_n\geq\chi_{1,0.05}^2)$

где $\chi_{1,0.05}^2$ это  1-0.05 процентиль , что можно найти в таблице.

### Likelihood ratio test

$2(\ell(X_1,\dots,X_n;\hat\theta_{\text{MLE}})-\ell(X_1,\dots,X_n;\theta_0))\xrightarrow[(d)]\chi^2_1$

и тест будет выглядит как 

$\psi'=\mathbb{1}(T_n'\geq\chi_{1,0.05}^2)$

# Подсчитаем тесты

- Wald's test: $\psi=\mathbb{1}(T_n>3.84)$
- Likelihood ratio test: $\psi'=\mathbb{1}(T_n'>3.84)$

У обоих тестов одинаковые критические значения

Подсчитаем тесты при наблюдаемых данных $\hat\theta_{\text{MLE}}=2.45,\, n=100$
$H_0:\theta=2\\
H_1:\theta\neq2$

**Wald's test**

$T_n=n \dfrac{1}{\hat\theta_{\text{MLE}}}^2(\hat\theta_{\text{MLE}}-\theta_0)
=100\dfrac{1}{2.48^2}(2.45-2)^2=3.37$

$\psi=\mathbb{1}(T_n>3.84)=\mathbb{1}(3.37>3.84)=0$\
 **Failed to reject**

**Likelihood Ratio test**\
$T_n=2(\ell(X_1,\dots,X_n;\hat\theta_{\text{MLE}})-\ell(X_1,\dots,X_n;\theta_0))$\
$\ell(X_1,\dots,X_n;\theta_0)=n\log(\theta_0)+(-\theta_0-1)\sum_{i=1}^n\log X_i$ чтобы решить это уравнение нам нужно понять чему равно $X_i$? Но в наших данных нет $X_i$, но мы знаем чему равно $\hat\theta_{\text{MLE}}=\dfrac{n}{\sum_{i=1}^n\log X_i}$ из которого мы может выразить $\sum_{i=1}^n\log X_i=n\dfrac{1}{\hat\theta_{\text{MLE}}}$\
Итого получается $\ell(X_1,\dots,X_n;\theta_0)=n\log(\theta_0)+(-\theta_0-1)\dfrac{n}{\hat\theta_{\text{MLE}}}$ 

подставим значения

$T_n'=100\log(2)+(-2-1)\dfrac{100}{2.45}=3.85$ 

$\psi'=\mathbb{1}(T_n'\geq3.84)=\mathbb{1}(3.85\geq3.84)=1$

**Reject Null Hyp**


У нас два разных тест асимптотически сходящиеся к $\chi^2_1$ и оба основываются на MLE но дают **разные результаты**. возможно это удивительно, но в тоже время это очень разные тесты, основанные на разсных асимптотиках. И учитываю что наша выборка конечно, можно ожидать что они дадут разные ответы.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
# Генерируем значения для нормального распределения
df = 10
x = np.linspace(0, 40, 1000)
y = chi2.pdf(x, df=df)
```

```python
# Вычисляем критические значения для двухстороннего теста
alpha=0.05
z_crit = chi2.ppf(1 -alpha/2, df =df)

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=fr"Распределение $\chi^2_{{{df}}}$")
# точки на графике
# plt.scatter([z_crit,-z_crit],[0,0], color='magenta', label=r'$q_{\alpha/2}$')

# plt.text(0.65, 0.02, r'$q_{\alpha/2}='+str(z_crit)[:4]+'$', fontdict={"fontsize":'x-large'}, color='black')
# plt.text(-1.9, 0.02, r'$q_{\alpha/2}='+str(-z_crit)[:5]+'$', fontdict={"fontsize":'x-large'}, color='black')
# Заполняем области отклонения
plt.fill_between(x, 0, y, where=(x < -z_crit)|(x >= z_crit), color='magenta', alpha=0.4, label="Область отклонения 5%")


# результатыт теста 
wald_test = -2.04
# plt.scatter([wald_test],[0], color='black', label=r'$W=-2.04$')
# plt.fill_between(x, 0, y, where=(x < wald_test), color='black', alpha=0.7, label="Область Теста")

# считаем значение p-value для теста
p_value = 2*chi2.cdf(-2.04, df=df)

p_value

# Добавляем метки и легенду
plt.title("Область отклонения для одностороннего теста")
plt.xlabel("Значения статистики теста")
plt.ylabel("Плотность вероятности")
plt.legend()

# Показываем график
plt.grid(True)
plt.show()
```
