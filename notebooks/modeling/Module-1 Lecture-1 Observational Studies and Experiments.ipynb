Рассмотрим главную задачу - определить, эффективны ли маммографии в предотвращении смертности от рака груди. Эксперимент рассматривает это в терминах оценки влияния одной переменной на другую. В самой простой постановке есть две переменные:

- **Переменная лечения**, также называемая **независимой переменной**, - это то, что мы можем изменять и что вызывает изменения других переменных.

- **Результирующая переменная**, также называемая **зависимой переменной**, - это то, что мы наблюдаем и что зависит от переменной лечения.

Цель экспериментальных и статистических процедур - установить связь между переменной лечения и результирующей переменной. Весь процесс мы называем **экспериментальным дизайном**.

**Переменные лечения и результата в исследовании маммографии**

В исследовании маммографии мы определяем, что такое **переменная лечения** и что такое **результирующая переменная**. Начнем с результирующей переменной, так как это проще.

- Для *результирующей переменной* мы можем установить, **является ли данное лицо (в исследовании) умершим от рака груди**. Это тот реальный результат, который мы намерены предотвратить. Это также определенное количество (как в агрегированном, так и в единичном лечении).

- Для *переменной лечения* естественным первым ответом является, проходил ли конкретный человек маммографию. Однако функционально невозможно напрямую применить «лечение» в виде прохождения маммографии, так как человек может просто отказаться. Поэтому мы определяем **переменную лечения** как **предложение маммографии**.

Тщательно определяя переменные лечения и результата в этом исследовании, мы можем точно указать, что будет анализировать это исследование. Конкретные определения этих переменных также будут направлять процедуру сбора и анализа данных.

**Проверка гипотез**

Суть проверки гипотез можно свести к ситуации, когда у нас есть утверждение и соответствующие данные, которые могут помочь нам оценить его точность. Короче говоря, мы хотим ответить на вопрос, поддерживают ли данные это утверждение или нет. Утверждение называется «гипотеза», и процедура называется «проверка гипотез».

**Идеально, если у нас есть доступ ко всем возможным данным во всех возможных исходах, мы можем свести это к арифметическому вычислению. В более реалистичной ситуации, однако, у нас есть только часть популяции, с которой мы работаем. Это создает элемент неопределенности, так как мы не можем контролировать, какой именно образец мы получаем, и это дает повод для статистики.**

Мы начнем с рассмотрения **процесса сбора данных**, который представляет собой процесс генерации этих выборок. Это часть более широкой концепции экспериментального дизайна, который также охватывает процедуру, когда мы используем собранные данные для проведения соответствующих выводов о поставленных гипотезах.

# 5. Study considerations and properties

**Некоторые свойства эксперимента, которые следует учитывать**

Из нашего текущего набора данных обсудим несколько важных моментов при разработке нашего исследования.

- Отбор пациентов: Некоторые популяции более склонны к развитию рака груди, чем другие, в зависимости от их предыдущего состояния здоровья. Поэтому интерпретация любых полученных результатов будет зависеть от того, как мы определили нашу процедуру отбора. В общем, то, как мы выбираем группы лечения и контроля, повлияет на популяцию, для которой вывод будет действителен.

- Контрольная группа: Нам нужно сравнить результирующую переменную для тех, кто получил лечение, с базовым уровнем (т.е. с контрольной группой). Здесь контрольная группа (которой не предлагали маммографию) должна быть **сопоставима** с группой лечения (которой предлагали маммографию).

- Характеристики пациентов: Один из способов сделать точное сравнение и интерпретацию результатов - это обеспечить, чтобы группа лечения была репрезентативной по таким факторам, как состояние здоровья, возраст и этническая принадлежность, и была похожа по этим параметрам на контрольную группу. Таким образом, мы можем приписывать любые различия в результатах лечению, а не различиям в ковариатах. В общем исследовании исследователи сами определяют, какие характеристики учитывать.

Подход, который учитывает вышеупомянутые три пункта, называется **рандомизированным контролируемым испытанием (РКИ)**.

# 6. Randomized Controlled Trials (RCT)


**Рандомизированное контролируемое испытание**

**Рандомизированное контролируемое испытание (РКИ)** - это экспериментальный дизайн, в котором лечение назначается случайным образом.

Мы начинаем с достаточно большого **набора людей (или "единиц")**, делим их случайным образом на группу лечения и контрольную группу. (В языке экспериментального дизайна каждая различимая личность, для которой можно уникально определить принадлежность к группе лечения или контроля, называется единицей.)

Лечение применяется к единицам лечения, в то время как контрольные единицы либо просто оставляются для наблюдения, либо **получают плацебо**. Затем, по закону больших чисел, мы ожидаем, что разница в средних значениях любой соответствующей характеристики между двумя группами будет довольно мала. Это позволяет нам взять разницу средних результирующих переменных в двух группах, чтобы оценить эффект лечения.

РКИ очень распространены в медицине для оценки нового медицинского продукта, такого как лекарство или вакцина. Самый простой случай РКИ - это испытание лекарства, когда пациенты с конкретным заболеванием случайным образом распределяются на группу лечения или контрольную группу. Пациенты в группе лечения получают новое изучаемое лекарство, в то время как пациенты в контрольной группе получают плацебо. Это плацебо может быть либо существующим стандартным лечением, либо лекарством без медицинского эффекта (например, сахарной таблеткой). Затем мы можем позже сравнить результаты пациентов, получивших новое лекарство, с результатами тех, кто получил плацебо.

**Стратификация в рандомизированном контролируемом испытании**

Одна из распространенных модификаций РКИ - **использование стратификации**. Здесь мы заранее делим популяцию на группы, а затем выбираем пропорционально из каждой группы. Это позволяет не оставлять беспристрастность нашей выборки в отношении конкретной классификации или характеристики на волю случая. Стратификация также позволяет проводить **анализ подгрупп**, то есть анализировать эффект лечения в пределах конкретной группы.

В сценарии испытания лекарства распространенная стратификация проводится путем разделения пациентов на группы в соответствии с определенными демографическими категориями. Это связано с различиями в многих физиологических функциях среди групп разного возраста, пола или этнической принадлежности. Случайное распределение затем может быть проведено внутри каждой группы, что позволяет проводить валидные выводы как в каждой подгруппе, так и в целом, путем агрегирования результатов по группам.

**Проблемы размера выборки**

В анализах подгрупп и в РКИ, где наша выборка мала, однако, мы, вероятно, столкнемся с проблемами размера выборки, что может повлиять на точность оценки. Когда мы берем выборку из небольшого количества точек, более вероятно, что мы получим либо большое количество подряд больших точек, либо подряд малых точек, по сравнению с большой выборкой. Таким образом, разница в средних результатах будет в некотором смысле означать меньше в небольшой выборке. Нам нужна относительно большая выборка, чтобы обнаружить небольшой эффект лечения.








# 7. Ethical and Human Considerations in a Study


В исследовании мы должны быть осторожны с человеческими факторами, так как их игнорирование может вызвать предвзятости во всем эксперименте. Такие предвзятости могут привести к ошибкам в причинно-следственном рассуждении, что может сделать статистические выводы исследования недействительными. В процессе академического рецензирования тщательно проверяется, как предлагаемая методология устраняет такие предвзятости. Кроме методологических вопросов, этические соображения также будут определять выбор процедур, выполняемых в ходе исследования.

**Двойное слепое исследование**

В любом эксперименте, включающем людей, факторы, связанные с человеческим поведением, могут влиять на результаты, скрывая эффекты лечения. Например, если пациенты в испытании лекарства узнают, что они действительно получили новое лекарство, их поведение может измениться разными способами, например, они могут стать более или менее осторожными в отношении своих решений, связанных со здоровьем. Такие изменения очень трудно смоделировать, поэтому мы стремимся минимизировать их влияние настолько, насколько это возможно.

Стандартный способ решения этой проблемы - **проведение двойного слепого исследования**, также называемого **слепым экспериментом**. Здесь людей лишают возможности знать, находятся ли они в группе лечения или контроля. В то же время, те, кто отвечает за эксперимент, и любой другой, кто может взаимодействовать с пациентом, также не должны знать, находится ли пациент в группе лечения или контроля. Это необходимо для предотвращения различных когнитивных предвзятостей, таких как [предвзятость наблюдателя](https://en.wikipedia.org/wiki/Observer_bias) или [предвзятость подтверждения](https://en.wikipedia.org/wiki/Confirmation_bias), которые могут повлиять на эксперимент.

В некоторых случаях невозможно обеспечить *полностью двойное слепое исследование*. Например, в исследовании маммографии пациенты однозначно будут знать, проходили ли они маммографию. Если мы изменим лечение на предложение пройти маммографию (которое они могут отклонить), тогда мы не имеем и не хотим двойной слепоты. Здесь решение пациента о прохождении маммографии после предложения и его/ее последующее поведение является частью оцениваемого нами причинного эффекта.

**Этические соображения и согласие**

В этом примере, частично из-за этических соображений, мы решили, что нашим лечением будет предложение пройти маммографию, а не проведение самой маммографии. Предположим, однако, что мы намерены оценить причинный эффект фактического проведения маммографии. Тогда один из вариантов - провести большое количество маммографий и затем случайным образом отбрасывать половину из них (без дальнейших действий).

Однако этот вариант вызывает этические сомнения, так как предоставление половине пациентов, возможно, ложной информации об их риске рака груди опасно. Любое исследование с участием людей имеет связанные с этим этические соображения, и все такие исследования подлежат рассмотрению соответствующим этическим комитетом.

Кроме того, **согласие** является важной составляющей современных этических стандартов, что означает, что требуется, чтобы участники исследования понимали его параметры и соглашались с его условиями. Любой, кто участвует в эксперименте или исследовании, должен быть проинформирован о его рисках и возможных последствиях. (Например, см. этот [справочник о согласии в экспериментальном исследовании](https://umc.edu/Research/Research-Offices/Clinical-Trials/Researchers/Road-Map/Study-Assessment/Design-ICF.html).)


# 8. Randomization, and Control Variables


**Рандомизация**

Рандомизированное контролируемое испытание (РКИ) позволяет нам выявить причинный эффект, обеспечивая, что в ожидании группы лечения и контроля будут одинаково распределены по любым соответствующим характеристикам (или по любой характеристике). Исследование маммографии является таким примером. Здесь мы можем взять большую выборку из 62 000 взрослых женщин, застрахованных по определенному плану, предложить маммографию 31 000 из них (лечение) и не предлагать оставшимся 31 000 (контроль).

В зависимости от пары переменных лечения и результата может быть невозможно использовать РКИ для выявления причинного эффекта. Предположим, исследователь хочет оценить влияние курения на заболеваемость раком легких. Мы не можем напрямую изменить поведение человека в отношении курения, поэтому мы не можем сформулировать РКИ для этого причинного эффекта. Однако, если мы изменим лечение на то, что мы можем назначить и, следовательно, рандомизировать, например, давать ли человеку советы о вреде курения, тогда мы можем оценить причинный эффект.

В ситуациях, когда рандомизация невозможна, нашим лучшим вариантом является **наблюдательное исследование**, где мы пытаемся оценить причинный эффект на основе существующих данных. Например, при изучении влияния определенных генов у людей, редактирование генов считается неэтичным и поэтому не проводится. Существуют методы, такие как инструментальные переменные, которые позволяют нам использовать причинный эффект от одного лечения для оценки другой связи, которая нас больше интересует. Эти методы, однако, выходят за рамки данного курса. (Курс [14.310x Data Analysis for Social Scientists](https://mitxonline.mit.edu/courses/course-v1:MITxT+14.310x/) дает довольно подробный обзор этой техники.)








# 9. Observational study: confounding and control variables

Теперь, что мы можем сделать, если рандомизация невозможна? Это означает, что с точки зрения методов мы ограничены **наблюдательным исследованием**, где исследователи не применяют лечение, а только наблюдают. Сложность в этом сценарии заключается в том, что маловероятно, что это лечение будет рандомизировано.

**[Смешивающие/Спутывающие переменные](https://ru.wikipedia.org/wiki/%D0%A1%D0%BF%D1%83%D1%82%D1%8B%D0%B2%D0%B0%D1%8E%D1%89%D0%B0%D1%8F_%D0%BF%D0%B5%D1%80%D0%B5%D0%BC%D0%B5%D0%BD%D0%BD%D0%B0%D1%8F)**

Поскольку лечение не рандомизировано, мы не можем оценить причинный эффект, просто взяв разницу в средних значениях характеристик пациентов. Смещение в разнице средних возникает, когда даже без лечения у единиц, которые с большей вероятностью получат лечение, базовый результат отличается от тех, у кого вероятность лечения ниже. В результате мы не можем определить, является ли разница следствием лечения или различиями в базовых значениях.


**Пример:** Предположим, мы хотим оценить эффект специального обучения на успехи студентов. Если студенты, которые записываются на специальное обучение, **уже более мотивированы и имеют лучшие начальные академические показатели по сравнению с теми, кто не записывается**, то разница в успеваемости между этими группами будет искажена. Даже если специальное обучение окажет положительное влияние, сложно определить, какая часть улучшения связана с обучением, а какая — с исходными характеристиками студентов.

**Пример:** Предположим, мы хотим оценить **влияние нового лекарства** на **снижение уровня холестерина**. Если пациенты, которым назначают это лекарство, **уже ведут более здоровый образ жизни** и имеют более низкий исходный уровень холестерина по сравнению с теми, **кому лекарство не назначают**, **то разница в уровнях холестерина между этими группами будет искажена**. Даже если лекарство оказывает положительный эффект, сложно определить, какая часть улучшения связана с лекарством, а какая — с исходными характеристиками и образом жизни пациентов.

Это явление называется **Смешивание/Спутывание (Counfounding)**, и любая внешняя переменная, которая может влиять как на лечение, так и на результирующие переменные, называется **смешивающей переменной**. Тоесть она влияет на зависимую так и на независимую переменные, результатом чего является ложная зависимость.

Спутывание - это причинная концепция, элемента [казуальной модели](https://ru.wikipedia.org/wiki/%D0%9A%D0%B0%D1%83%D0%B7%D0%B0%D0%BB%D1%8C%D0%BD%D0%B0%D1%8F_%D0%BC%D0%BE%D0%B4%D0%B5%D0%BB%D1%8C), и как таковая она не может быть описана в терминах корреляций или ассоциаций.

> Пусть $P(y|\text{do}(x))$ вероятность события $Y=y$ при гипотетическоем вмешательстве $X=x$. $X$ и $Y$ не спутываются тогда и только тогда, когда выполняется следующее условие
> $$P(y|do(x))=P(y|x)$$
> для всех значений вероятностей события $X=x$ и события $Y=y$, где $P(y|x)$ - это условная вероятность $X=x$
> 
>В принципе, определяющее равенство ${\displaystyle P(y\mid {\text{do}}(x))=P(y\mid x)}$ можно проверить на основе модели генерации данных, предполагая, что у нас есть все уравнения и вероятности, связанные с моделью. Это делается путём моделирования вмешательства ${\displaystyle {\text{do}}(X=x)}$ (см. [Байесовская сеть](https://ru.wikipedia.org/wiki/%D0%91%D0%B0%D0%B9%D0%B5%D1%81%D0%BE%D0%B2%D1%81%D0%BA%D0%B0%D1%8F_%D1%81%D0%B5%D1%82%D1%8C)) и проверки того, равна ли результирующая вероятность Y условной вероятности ${\displaystyle P(y\mid x)}$. При этом оказывается, что свойств графа достаточно для проверки равенства $ {\displaystyle P(y\mid {\text{do}}(x))=P(y\mid x)}$.

Подход стратификации, о котором мы говорили ранее, является **возможным** решением этой проблемы. Если смешивающая переменная - это демографическая категория, то балансировка переменной лечения в каждой категории решает проблему. В примере с курением, смешивающая переменная заключается в том, что люди старшего возраста как курят больше, так и более подвержены раку легких. Тогда выборка одинакового числа курящих и некурящих в каждой возрастной группе исправляет это смешение.

**Пример с лекарством и холестерином:** Для решения проблемы смешивания мы можем **стратифицировать выборку по уровням физической активности.** Например, мы можем разделить всех участников на группы с **низким, средним и высоким уровнем физической активности**. В каждой группе мы будем сравнивать уровни холестерина у тех, кто получает новое лекарство, и у тех, кто его не получает. Это позволит минимизировать влияние физической активности на результаты и более точно оценить эффект самого лекарства.

**Пример с курением и возрастом:** В примере с курением, смешивающая переменная заключается в том, что люди старшего возраста как курят больше, так и более подвержены раку легких. Тогда **выборка одинакового числа курящих и некурящих в каждой возрастной группе исправляет это смешение.** Например, мы можем разделить всех участников на возрастные группы (например, 20-30, 31-40, 41-50 и т.д.) и в каждой возрастной группе уравновесить количество курящих и некурящих. Это позволит устранить влияние возраста на результаты и более точно оценить эффект курения на риск развития рака легких.

**Использование контрольных переменных**

Использование **контрольных переменных** - еще одно решение этой проблемы. Изучение контрольных переменных в контексте **причинности** сложно, поэтому мы кратко рассмотрим, что делает контрольную переменную подходящей. В идеальной ситуации мы выбираем контролируемые параметры, которые учитывают все возможные **источники смещения, факторы, которые влияют как на лечение, так и на результирующие значения**. Однако большинство исследований идентифицируют только часть этих источников. После выявления контрольных параметров существует множество техник для **учета их влияния**, наиболее распространенной из которых является **мультивариативная регрессия**.

**Пример 1:** Исследование, изучающее **влияние приема нового препарата на артериальное давление**. В этом случае, **возраст пациентов и индекс массы тела (ИМТ) могут служить контрольными переменными**. Пожилые пациенты и пациенты с высоким ИМТ могут иметь более высокое артериальное давление, независимо от приема препарата. Поэтому для корректного анализа эффекта препарата, важно учитывать влияние этих переменных.

**Пример 2:** Исследование, оценивающее **влияние диеты на уровень сахара в крови у пациентов с диабетом**. **Здесь контрольными переменными могут быть уровень физической активности и генетическая предрасположенность**. Оба этих фактора могут влиять на уровень сахара в крови, независимо от диеты. Учет этих переменных с помощью мультивариативной регрессии поможет более точно определить влияние диеты.

**Пример 3:** В исследовании, анализирующем влияние хирургического вмешательства на восстановление после травмы, контрольной переменной может быть общее состояние здоровья пациента. Пациенты с лучшим общим состоянием здоровья могут быстрее восстанавливаться, независимо от проведенной операции, поэтому состояние здоровья необходимо учитывать при анализе результатов лечения.

Например, рассмотрим исследование, которое изучает влияние лет обучения на доход в возрасте 30 лет. Мы ожидаем, что человек из обеспеченной семьи будет дольше оставаться в школе по многим причинам, одной из которых является высокая стоимость высшего образования. В то же время, социально-экономический статус человека в детстве влияет на будущий доход по причинам, не связанным с образованием, например, через более высокие ожидания и лучшие неформальные сети. Таким образом, доход семьи является подходящей контрольной переменной.

**Проблемы с контрольными переменными и причинностью**

Теперь рассмотрим, что может сделать контрольную переменную неподходящей. Рассмотрим исследование, которое изучает влияние курения на частоту инсультов. Затем предположим, что госпитализация предлагается в качестве контрольной переменной, что означает, что, например, мы проводим наблюдательное исследование с пациентами в определенной больнице как нашей выборкой. В этом случае контроль может на самом деле ввести смещение, потому что как курение, так и инсульт могут вызвать госпитализацию, хотя первое - более косвенно. Это отрицательно скажется на оценке взаимосвязи, потому что, если мы учитываем госпитализированных пациентов, **эффект одной причины будет вытеснять эффект другой.**


*Причинная диаграмма курения, инсульта и госпитализации*

Чтобы лучше понять, почему контроль по общей переменной-потомку в причинной диаграмме приведет к смещению выводов, рассмотрим еще один более простой пример:

Представьте, что $X$ и $Y$ — это два независимых события, например, подбрасывание двух монет. $Z$ — это результат логической операции ИСКЛЮЧАЮЩЕЕ ИЛИ (XOR), которая принимает значение 1, когда только одно из двух событий происходит (то есть одна монета выпала орлом, а другая решкой), и значение 0, когда оба события либо оба орлы, либо оба решки.

Когда мы пытаемся выяснить, зависят ли $X$ и $Y$ друг от друга, мы можем посмотреть на множество результатов подбрасывания монет и увидеть, что они действительно независимы. Но если мы будем учитывать только те случаи, когда $Z = 1$ (то есть одна монета выпала орлом, а другая решкой), мы неожиданно заметим, что $X$ и $Y$ кажутся зависимыми. Когда $Z = 1$, знание результата одной монеты дает нам информацию о результате другой монеты. Например, если мы знаем, что $X = 0$ (решка), то $Y$ обязательно должно быть 1 (орел), чтобы $Z$ было равно 1. Это создает ложное впечатление, что $X$ и $Y$ связаны, хотя на самом деле они независимы.

Этот пример показывает, как выбор неправильной контрольной переменной (в данном случае $Z$) может создать иллюзию зависимости между независимыми переменными. Контролируя такую переменную, мы можем ввести смещение в наши выводы и получить неверное представление о взаимоотношениях между переменными.

пример причинного гафа, на нем мы обозначает направленным ребром - влияние одной переменной на другую.


Вспомним что Confounder это перменные которыя влияют как на зависимые так и не зависимые переменные. 

**к примеру**, нас интересует эффект $X$ на $Y$. Какие из других перменных являются Спутывающие для перменных $X$ и $Y$?

По определению Спутывающие это те что влияют на $X$ и $Y$. Так на нашем графе $A$ влияет на $X$ и $Y$. А так же $E$ влияет на $X$ через $B$, и На $Y$ через $C$. Получается что $A$ и $E$ это спутывающие перменные.

**Предположем мы хотим оценить прямую зависимость $X$ и $Y$ от данных. Что мы должны делать?**

Чтобы верно оценить если есть зависимость между $X$ и $Y$, мы должны "блокировать" все пути которы могут показать зависимость между $X$ и $Y$. Даже если не существует зависимости между $X$ и $Y$ на самом деле, то общая причинность (спутывающая перменная) может сделать $X$ и Y ассоциированными. Тогда $A$ и $E$ нужно контролировать. 
Если нет зависимотси между $X$ и $Y$, то переменная $D$ может также сделать их подобно зависимым одной от другой. Так что $D$ тоже нужно контролировать.


Особое внимание должно удивляться $F$ - общий потомком от $X$ и $Y$. Когда нет связи между $X$ и $Y$, спутывающая пемернная $F$ может фактически сделать $X$ и $Y$ связанными. К примеру - допустим что $X$ и $Y$ независимы. а перменная $F=X \vee Y$. Если перменная $F=1$ и при этом $X=0$, мы точно знаем что $Y=1$ $ X \not\! \perp \! \! \! \perp Y | F$ - X не является условно независимым от Y при условии F. Поэтому не стоит учитывать $F$

После учета переменных $A$, $E$ и $D$ оценка будет правильной.

Учитывать $B$ и $C$ не обязательно, так как мы уже учли $E$, и путь $X$-$B$-$E$-$C$-$Y$ заблокирован. Но учет $B$ и $C$ тоже не причинит других проблем. Поэтому последний вариант тоже приемлем.



> Это уравнение можно прочитать и понять следующим образом:
> 
> "X не является условно независимым от Y при условии F."
> 
> В математической статистике и теории вероятностей этот символ $ \not\! \perp \! \! \! \perp $ используется для обозначения зависимости, противоположной условной независимости. 
> 
> Чтобы разобрать это более подробно:
> 
> - $ X $ и $ Y $ — это две случайные величины или переменные.
> - $ F $ — это третья переменная или множество переменных, при условии которых мы рассматриваем зависимость между $ X $ и $ Y $.
> 
> Если бы уравнение было записано как $ X \perp \! \! \! \perp Y | F $, это означало бы, что $ X $ и $ Y $ являются условно независимыми при условии $ F $. То есть знание $ F $ делает $ X $ и $ Y $ независимыми друг от друга.
> 
> Но в данном случае, поскольку уравнение имеет вид $ X \not\! \perp \! \! \! \perp Y | F $, это означает, что даже при знании $ F $ случайные величины $ X $ и $ Y $ остаются зависимыми друг от друга.
> 
> Таким образом, уравнение указывает на наличие зависимости между $ X $ и $ Y $, несмотря на то, что мы знаем $ F $.
