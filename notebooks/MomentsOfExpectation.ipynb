```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.random.normal(size = (100000,))
print('E[Z]',np.mean(data))
print('E[Z2]',np.mean(data**2))
print('E[Z3]',np.mean(data**3))
print('E[Z4]',np.mean(data**4))
```

```python
plt.hist((data-0)/1, label='E[Z]', bins=100)
plt.show()
```

```python
plt.hist(data**2, label='E[Z2]', bins=100)
plt.show()
plt.hist(data**3, label='E[Z3]', bins=100)
plt.show()
plt.hist(data**4, label='E[Z4]', bins=100)
plt.show()
```
