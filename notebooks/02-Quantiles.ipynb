Что же такое квантильная функция?

допустим у нас есть переменная $X$ у которой CDF $F_X:\mathbb{R}\mapsto[0,1]$\
$F$ непрерыная и постоянно возрастающая функция. ее домен ограничен \[0,1\]

тогда для любого значение x можно сказать что 
$$\mathbb{P}(X<x) = F_X(x)$$

$$\mathbb{P}(X<1.1) = F_X(1.1)$$

если у нас есть $a, b \in [0,1], a>b$ тогда:

$$\mathbb{P}(a<X<b) = F_X(b)-F_X(a)$$
$$\mathbb{P}(|X|>a) = 2\mathbb{P}(X>a) = 2(1 - F(a))$$

# Quantiles

Допустим $\alpha\in(0,1)$. 

Квантиль порядка $1-\alpha$ случайной переименной $X$ это число $q_\alpha$ такое, что:
$$\mathbb{P}(X\leq q_\alpha)=1-\alpha$$

**90ый процентиль** получается при  $\alpha=0.1$
$$\mathbb{P}(X\leq q_{0.1})=1-0.1=0.9=90\%$$

$q_{0.1}$ - это значение **90го процентиля**

- $F(q_\alpha)=1-\alpha\,\quad$ CDF
- if $F$ is invertable $q_\alpha=F^{-1}(1-\alpha)$ смотреть на таблицу - наоборот
- $\mathbb{P}(X>q_\alpha ) = \alpha$ a.k.a **$\quad\alpha$ critical value**
- if $X=Z\sim\mathcal{N}(0,1): \mathbb{P}(|X|> q_{\alpha/2} )=\alpha$
| $\alpha$ | 2.5% | 5% | 10% |
|:--:|:--:|:--:|:--:|
| $q_\alpha$ | 1.96 | 1.65 | 1.28 |

$F^{-1}(p) = \inf  \big\{x \in \mathbb{R}: F(x) \ge p \big\}.$

```python

# Определяем параметры распределения и области отклонения
mu = 0  # среднее значение
sigma = 1  # стандартное отклонение
alpha = 0.05  # уровень значимости
# Генерируем значения для нормального распределения
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.cdf(x, mu, sigma)

# Вычисляем критические значения для двухстороннего теста
z_crit = norm.ppf(1 - alpha/2)

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="CDF F")

# Заполняем области отклонения
# plt.fill_between(x, 0, y, where=(x <= -z_crit) | (x >= z_crit), color='red', alpha=0.5, label="Область отклонения")


x_1 = 1.1
y_1 = norm.cdf(x_1, mu, sigma)

plt.vlines(x=1.1, ymin= 0, ymax=y_1, color='black', linestyles='dashed')
plt.hlines(y=y_1, xmin=-4, xmax=x_1, color='black', linestyles='dashed')
plt.scatter(x=[-4], y=[y_1], color='black')
plt.scatter(x=[x_1], y=[y_1], color='black')
plt.scatter(x=[x_1], y=[0], color='black')


# Добавляем метки и легенду
plt.title("Область отклонения для двухстороннего теста")
plt.xlabel("CDF")
plt.ylabel(r"Вероятность $X<x$")
plt.legend()
plt.xlim((-4.15, 4.15))
plt.ylim((-0.025, 1.025))
plt.arrow(x=-4, y=0, dx=0, dy=1, color='black', head_length=0.25/4 ,head_width =0.08, length_includes_head =True)
plt.arrow(x=-4, y=0, dx=8, dy=0, color='black', head_length=0.25, head_width =0.02, length_includes_head =True)
#  fontsize or size: float or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
plt.text(-3.8, 0.9, r'$\mathbb{P}\,(X<x$)', fontdict={"fontsize":'x-large'})
plt.text(1.5, 0.8, r'$F(x)$', fontdict={"fontsize":'x-large'})
plt.text(1.25, 0.05, r'$x=1.1$', fontdict={"fontsize":'x-large'})
plt.legend()
plt.box(on=None)
# Показываем график
# plt.grid(True)
plt.show()
```
